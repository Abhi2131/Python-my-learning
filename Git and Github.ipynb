Version control system have two main types:
        1. Centralised Version Control System (CVCS)
        2. Distributed Version Control System  (DVCS)

1. CVCS:
    - all local systems are connected directly with the central system eg.clearcse and subversion
      Disadv:
     - 1.if CVSC downs/loss files, all local system has to wait till central CVCS correct itself.
     - 2.no backup server if goes down or it crashes
     - 3.slower as compared to DVCS
     
-2. DVCS:
    - central system as well as all ocal systems are internally connected eg.Git and Github
      adv:
     - 1.any change made by one person is available for everyone connected in system
     - 2.overcomes all disadv of CVCS
     - 3.can backup code easily if server crashes or goes down
